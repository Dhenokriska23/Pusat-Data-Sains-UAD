#Avocado Prices



Tabel di bawah ini mewakili data pemindaian ritel mingguan 2018 untuk volume (unit) dan harga ritel Nasional. Data pemindaian ritel berasal langsung dari mesin kasir pengecer berdasarkan penjualan eceran sebenarnya dari alpukat Hass. Mulai tahun 2013, tabel di bawah ini mencerminkan kumpulan data ritel multi-outlet yang diperluas. Pelaporan multi-outlet mencakup agregasi dari saluran berikut: grosir, massa, klub, obat-obatan, dolar, dan militer. Harga Rata-rata (alpukat) di tabel mencerminkan biaya per unit (per alpukat), bahkan ketika beberapa unit (alpukat) dijual dalam kantong. Kode Pencarian Produk (PLU) di tabel hanya untuk alpukat Hass. Varietas alpukat lainnya (misalnya kulit hijau) tidak termasuk dalam tabel ini.

## Data Preparation

```
import pandas as pd
import numpy as np

avocado_df = pd.read_csv('/content/avocado.csv.zip')
#Keterangan isi index
print(avocado_df.keys())
avocado_df
```

```
avocado=pd.DataFrame(avocado_df[['4225','region']])
avocado.sort_values(by=["region"])
```

## Data Visualization

```
import plotly.express as px
import plotly.graph_objects as go
```

```
gbr_avocado = px.bar(avocado,
               x='region',
               y='4225')

gbr_avocado.update_layout(
    title='Diagram Batang banyak penjualan avocado ',
    title_font_color='blue',
    title_font_size=30,
    xaxis=dict(
        title='region',
        titlefont_size=16,
        tickfont_size=14),
    yaxis=dict(
        title='4225',
        titlefont_size=16,
        tickfont_size=14))

gbr_avocado.show()
```
